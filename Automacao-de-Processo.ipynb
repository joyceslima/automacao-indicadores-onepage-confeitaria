### Passo 1 - Importar Arquivos e Bibliotecas

```python
import pandas as pd
import pathlib
import win32com.client as win32
```

```python
emails = pd.read_excel(r'Bases de Dados\Emails.xlsx')
lojas = pd.read_csv(r'Bases de Dados\Lojas.csv', encoding='latin1', sep= ';') #encoding- algum caracter deve ser especial(acento ou ~); separador era (;)
vendas = pd.read_excel(r'Bases de Dados\confeitaria_detalhada.2.xlsx')
```

### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
#juntar as bases

vendas = vendas.merge(lojas, on= 'ID Loja')
display(vendas)
```

```python
print(vendas.dtypes)
```

```python
dicionario_lojas = {}

for loja in lojas['Loja']: #esse for é para conseguir procurar as vendas de cada loja
    dicionario_lojas[loja] = vendas.loc[vendas['Loja']==loja, :]

display(dicionario_lojas['Passei das Águas Shopping'])
```

```python
display(dicionario_lojas['Shopping Morumbi'])
```

```python
display(dicionario_lojas['Ribeirão Shopping'])
```

```python
#convertendo o formato para datetime
vendas['Data'] = pd.to_datetime(vendas['Data'])
```

```python
#para saber qual o dia mais recente (último dia de venda)
dia_indicador = vendas['Data'].max()
print(dia_indicador)
print('{}/{}'.format(dia_indicador.day, dia_indicador.month))
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
# identificar se existe a pasta das lojas
caminho_backup = pathlib.Path(r'Backup Arquivos Lojas')

arquivos_pasta_backup = caminho_backup.iterdir()

lista_nomes_backup = [arquivo.name for arquivo in arquivos_pasta_backup]

#criando as pastas de cada loja
for loja in dicionario_lojas:
    if loja not in lista_nomes_backup:
        nova_pasta = caminho_backup / loja
        nova_pasta.mkdir()
    
    #salvar dentro da pasta de cada loja
    nome_arquivo = '{}_{}_{}.xlsx'.format(dia_indicador.day, dia_indicador.month, loja)
    print(nome_arquivo)
    local_arquivo = caminho_backup / loja / nome_arquivo

    dicionario_lojas[loja].to_excel(local_arquivo)
```

### Passo 4 - Calcular o indicador para 1 loja

```python
loja = 'Shopping Recife'

vendas_shopp_recife = dicionario_lojas[loja]
vendas_shopp_recife_dia = vendas_shopp_recife.loc[vendas_shopp_recife['Data']==dia_indicador, :] #quero as linhas na coluna data, tem o dia indicador. Em todas as colunas

#calcular faturamento
faturamento_ano = vendas_shopp_recife ['Valor Final'].sum()
print(faturamento_ano)

faturamento_dia = vendas_shopp_recife_dia['Valor Final'].sum()
print(faturamento_dia)

#diversidade de produtos
qtde_produtos_ano = len(vendas_shopp_recife['Produto'].unique()) #unique 'retira' as duplicadas da coluna e mostra os únicos
print(qtde_produtos_ano)

qtde_produtos_dia = len(vendas_shopp_recife_dia['Produto'].unique())
print(qtde_produtos_dia)
```

```python
#calcular ticket médio ano
valor_venda = vendas_shopp_recife.groupby('Cupom Fiscal').sum(numeric_only=True) #numeric_only=True garante que apenas colunas numéricas sejam somadas
ticket_medio_ano = valor_venda['Valor Final'].mean()
print(ticket_medio_ano)

#calcular ticket médio dia
valor_venda_dia = vendas_shopp_recife_dia.groupby('Cupom Fiscal').sum(numeric_only=True)
ticket_medio_dia = valor_venda_dia['Valor Final'].mean()
print(ticket_medio_dia)
```
